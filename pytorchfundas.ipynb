Initializing Tensors

```python
import torch
my_tensor = torch.tensor([[1, 2, 3], [4, 5,6]], dtype=torch.float32,device="cpu", requires_grad=True)
print(my_tensor)
print(my_tensor.dtype)
print(my_tensor.device)
print(my_tensor.shape)
print(my_tensor.requires_grad)
```

Tensor Math & Comparison Operation

```python
x = torch.tensor([1, 2, 3])
y = torch.tensor([9, 8, 7])

#Addition
z1 = torch.empty(3)
torch.add(x, y, out=z1)

z2 = torch.add(x, y)
z = x + y

#Subtraction
z = x - y
# Division
z = torch.true_divide(x, y) #elemen wise if equal shape

#inplace ops
t = torch.zeros(3)
t.add_(x) #"_" in place ops
t += x # t = t + x

# Exponentation
z = x.pow(2)
print(z)
z = x ** 2

# simple comparision
z = x > 0
print(z) # element wise comparision

# Matrix Multiplication
x1 = torch.rand(2, 5)
x2 = torch.rand(5, 3)
x3 = torch.mm(x1, x2)
x3 = x1.mm(x2)

# Matrix exponentiation
matrix_exp = torch.rand(5, 5)
print(matrix_exp.matrix_power(3))

# Elemen-wise mul
z = x * y
print(z)

# dot product
z = torch.dot(x , y)
print(z)

# Batch Matrix Multiplication
batch = 32
n = 10
m = 20
p = 30

tensor1 = torch.rand(batch, n, m)
tensor2 = torch.rand(batch, m, p)
out_bmm = torch.bmm(tensor1, tensor2) # (batch, n, p)

#Example of Broadcasting
x1 = torch.rand(5, 5)
x2 = torch.rand(1 ,5)
z = x1 - x2
z = x1 ** x2

# other common tensor ops
sum_x = torch.sum(x, dim=0)
values, indices = torch.max(x, dim=0)
values, indices = torch.min(x, dim=0)
abs_x = torch.abs(x)
z = torch.argmax(x, dim=0)
z = torch.argmin(x, dim=0)
mean_x = torch.mean(x.float(), dim=0)
z = torch.eq(x, y)
sorted_y, indices = torch.sort(y, dim=0, descending=False)
z = torch.clamp(x, min=0)
x = torch.tensor([1, 0, 1, 1, 1], dtype=torch.bool)
z = torch.any(x)
z = torch.all(x) # False
```
